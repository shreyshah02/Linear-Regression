Importing the required libraries

```python
import numpy as np
from matplotlib import pyplot as plt
```

Loading the dataset food_truck_data.csv.text. 
Change the path if required

```python
data = np.loadtxt("food_truck_data.csv.txt", delimiter = ',')
print(data.shape)
population = data[:,0]
profit = data[:,1]
print(population.shape)
print(profit.shape)
```

```python
# d = np.loadtxt("food_truck_data.csv.txt", delimiter = ',')
population = population.reshape(-1, 1)
profit = profit.reshape(-1, 1)
print(population.shape)
print(profit.shape)
```

```python
X = np.empty((population.shape[0],2))
print(X.shape)
```

Creating the dataset matrix with 1st col. consisting of 1's corresponding to the bias and the 2nd column for population

```python
X[:,0] = 1
X[:,1] = population[:,0]
```

Generating the weights randomly of the shape 2 x 1

```python
W = np.random.random((X.shape[1],1))
print(W.shape)
```

Setting the learning rate to 0.02

Iterating through gradient descent until the change in the loss is less than 0.00001

```python
alpha = 0.02
loss_old = 0
loss = 10
i = 0
# for i in range(10000):
while abs(loss_old - loss)>0.00001:
    h = X.dot(W)
    loss_old = loss
    
    loss = np.sum(np.square(h - profit)) / (2*X.shape[0])
    print("Loss in iteration", i, '= ', loss)
    grad0 = np.sum(h - profit) / X.shape[0]
    grad1 = np.sum((h - profit)*population) / X.shape[0]
    W[0, 0] = W[0, 0] - alpha*grad0
    W[1, 0] = W[1, 0] - alpha*grad1
    plt.plot(i, loss, 'bo')
    i+=1
```

```python
print(W)
```

Plotting the actual data(red crosses) and the predicted model(blue line)

```python
plt.plot(population, profit, 'rx')

h = X.dot(W)
plt.plot(population,h)
plt.xlabel('population')
plt.ylabel('profit')
plt.show()
```

# Problem 2

Loading the data

```python
data_2 = np.loadtxt('housing_price_data.txt', delimiter = ',')
```

```python
print(type(data_2))
print(data_2.shape)
```

```python
house_size = data_2[:,0:1]
house_bed = data_2[:, 1:2]
house_price = data_2[:,2:3]
```

Scaling the features in the dataset

```python
size_mean = np.mean(house_size)
size_std = np.std(house_size)
house_size_norm = (house_size - size_mean)/size_std
```

```python
bed_mean = np.mean(house_bed)
bed_std = np.std(house_bed)
house_bed_norm = (house_bed - bed_mean)/bed_std

price_mean = np.mean(house_price)
price_std = np.std(house_price)
house_price_norm = (house_price - price_mean)/price_std
```

Creating the input data matrix

```python
X_new = np.empty_like(data_2)
X_new[:,0] = 1
X_new[:,1] = house_size_norm[:,0]
X_new[:,2] = house_bed_norm[:,0]
print(X_new)
```

Selecting Random weight vector

```python
W_new = np.random.random((X_new.shape[1], 1))
print(W_new.shape)
```

Setting the learning rate to 0.02 and performing the gradient descent algorithm for 1000 iterations

```python
alpha = 0.02
loss_old_new = 0
loss_new = 10
# i = 0
# while abs(loss_new - loss_old_new)>0.0001:
for i in range(1000):
    h_new = X_new.dot(W_new)
    loss_old_new = loss_new
    loss_new = np.sum(np.square(h_new - house_price_norm)) / (2*X_new.shape[0])
    print("Loss in iteration", i, '= ', loss_new)
    grad0 = np.sum(h_new - house_price_norm) / X_new.shape[0]
    grad1 = np.sum((h_new - house_price_norm)*house_size_norm) / X_new.shape[0]
    grad2 = np.sum((h_new - house_price_norm)*house_bed_norm) / X_new.shape[0]
    W_new[0, 0] = W_new[0, 0] - alpha*grad0
    W_new[1, 0] = W_new[1, 0] - alpha*grad1
    W_new[2, 0] = W_new[2, 0] - alpha*grad1
    plt.plot(i, loss_new, 'bo')
#     i+=1
```

```python
print(W_new)
```

Rescaling the output

```python
h_new = X_new.dot(W_new)
h_new = h_new*price_std + price_mean
```

Plotting the predicted price by the model against size of the house(x axis) and the no. of bed rooms(y axis)

```python
ax = plt.axes(projection = '3d')
print(house_size.shape, house_bed.shape, h_new.shape)
ax.scatter3D(house_size[:,0], house_bed[:,0], h_new[:,0])
ax.set_xlabel('House Size sq.feet')
ax.set_ylabel('No. of beds')
ax.set_zlabel('Predicted Price')
```

Precicting the price for a house with size 1800 sq. feet and 2 bedrooms. The 1st value is 1 corresponding to the bias.

```python
a[0,1] = (a[0,1] - size_mean)/size_std
a[0,2] = (a[0,2] - bed_mean)/bed_std
```

```python
pr_n = a.dot(W_new)
```
